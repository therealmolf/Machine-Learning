```python
import numpy as np 
from scipy.integrate import odeint
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def model(y, t):
    if t < 10.0:
        u = 0
    else:
        u = 2
    dydt = (-y + u) / 5
    return dydt
```

```python
y0 = 1
t = np.linspace(0,35)
```

```python
y = odeint(model, y0,t )
```

```python
plt.plot(t, y, "g:", label = "hello")
plt.xlabel("time")
plt.ylabel('y(t)')
plt.show()
```
